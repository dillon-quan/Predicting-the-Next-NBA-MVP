```python
import pandas as pd
df = pd.read_csv('Seasons_Stats.csv')
```

The code in the following cell was written to use player data from year 2000 and onward.

```python
df = df[df.Year >= 2000]
df.info()
```

```python
pd.options.display.max_columns = None
df.tail()
```

This cell here was to change certain features from object or float to int type.

```python
l = ['Year','Age', 'G', 'GS']
for feature in l:
    df[feature] = df[feature].astype(dtype ='int')
```

The following converts the basic stat line features from Total to per game stats of each player. The features that were changed are in list to_per_game.

```python
to_per_game = [ 'MP', 'FG', 'FGA', '3P', '3PA', '2P', '2PA','FT', 'FTA', 'ORB', 'DRB', 'TRB', 'AST', 'STL','BLK', 'TOV', 'PF', 'PTS']
for index,player in df.iterrows():
    for per_game in to_per_game:
        df.loc[index,per_game] = df.loc[index,per_game]/df.loc[index,'G']
```

```python
df = df.drop(['blanl','blank2'], axis=1)
```

```python
df.head()
```

```python
df['Year1'] = df.Year
```

The following cell first groups the dataframe by the year and name of the player and performs a count to discern players that were traded. To mask the dataframe, the index must match the filter so the index was set to do so. Lastly, the dataframe was set such that only players who have played for 1 team that season is left since players that are traded during a season are never MVPs.

```python
x = df.groupby(['Year','Player']).count()
df = df.set_index(['Year','Player'])
df = df[x.Tm == 1]
```

The following cell here creates a new column MVP for the dataset which contains the values 1 or 0 for MVP and non MVP respectively.

```python
df['MVP'] = 0
MVP = ["Shaquille O'Neal*", 'Allen Iverson*', 'Tim Duncan','Tim Duncan', 'Kevin Garnett', 'Steve Nash', 'Steve Nash', 'Dirk Nowitzki', 'Kobe Bryant', 'LeBron James','LeBron James','Derrick Rose', 'LeBron James','LeBron James','Kevin Durant','Stephen Curry','Stephen Curry','Russell Westbrook']
Season = [2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017]
MVP_winner = list(zip(MVP,Season))
for mvp, year in MVP_winner:
    df.loc[(year,mvp),'MVP'] = 1
    
df_mvps = df[df.MVP == 1]
df_non_mvps = df[df.MVP == 0]
df_mvps = df_mvps.drop('Year1', axis = 1)
```

The following cell will randomly sample 99 players from each year from the dataframe that are non_mvps. This newly random sample data will then be concatenated with the dataframe that contains the mvps to have a dataset that is randomly sampled with all MVPs from the year 2000 till present.

Problems:
-Each time I run the code, the sample is different. Is there a way to sample the same random data each time the code is written?

```python
df1 = pd.DataFrame()
for year in range(2000,2018):
    random_sample = df_non_mvps.loc[(year,slice(None))].sample(99, replace = False)
    df1 = pd.concat([df1,random_sample], axis = 0)
    
df1 = df1.reset_index()
df1 = df1.rename(columns={'Year1':'Year'})
df1= df1.set_index(['Player','Year'])
df_mvps = df_mvps.swaplevel()
df1 = pd.concat([df1, df_mvps], axis = 0)
df1.head()
```
