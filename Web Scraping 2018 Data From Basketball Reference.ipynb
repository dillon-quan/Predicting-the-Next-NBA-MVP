```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

pd.options.display.max_columns = None
```

```python
#a function that scraps the column headers from the url

def get_column_headers(url, n):
    """Returns the data table column headers from the webpage. The function takes two arguments: url of the webpage
        ,number of headers to be scrap."""
    
    #setup the get request and the parser that will be used to scrap the page
    r = requests.get(url)
    html_doc = r.text
    soup = BeautifulSoup(html_doc, 'html.parser')
    
    columns = []
    
    #iterating through the th tages for the headers.
    for i in range(1,n):
        header = soup.find('tr').find_all('th')[i].getText()
        columns.append(header)
    return columns
```

```python
#a function that scraps the player data from the given url

def get_player_data(url, m_players,stats_size):
    """Returns the player data from the webpage. The function takes three arguments: url of the webpage, 
        number of players, number of stats from each player."""
    
    #setup the get request and the parser that will be used to scrap the page
    r = requests.get(url)
    html_doc = r.text
    soup = BeautifulSoup(html_doc, 'html.parser')
    NBA_2018_players = []

    #iterating through td tags to get the data for each player 
    for i in range(1, m_players):
        player = []
        for k in range(stats_size):
            try:
                player.append(soup.find_all('tr')[i].find_all('td')[k].getText())
            except:
                continue
        NBA_2018_players.append(player)

    return NBA_2018_players
```

```python
#to retrieve the per_game stats of each player from 2018

url = 'https://www.basketball-reference.com/leagues/NBA_2018_per_game.html'
columns = get_column_headers(url, 30)
data = get_player_data(url, 662, 30)
df_per_game = pd.DataFrame(data, columns = columns)
df_per_game.head()
```

```python
#to retrieve the advance stats of each player from 2018

url = 'https://www.basketball-reference.com/leagues/NBA_2018_advanced.html'
columns = get_column_headers(url, 29)
data = get_player_data(url, 662, 29)
df_ad = pd.DataFrame(data, columns = columns)
df_ad.head()
```

```python
#drop redundant columns and empty rows and merge the two dataframes together.

df_ad = df_ad.drop(['Pos','Age','Tm','G','MP', '\xa0'], axis = 1)
idx = [index for index, player in df_ad.iterrows() if player[0] == None] 
df_ad = df_ad.drop(idx)
df = pd.merge(df_per_game,df_ad, how ='inner', on = ['Player'])
df.replace('', np.nan, inplace = True)
df.fillna(0, inplace = True)
```

```python
# import the Win/Lose Ratio for each player for 2018

df_team = pd.read_csv('2018_win_ratio.csv')
df_team = df_team.iloc[:, 2:]
df_team1 = df_team[['Team','Year', 'W/L%']]
df_team1 = df_team1.rename(columns={'Team':'Tm'})
df = pd.merge(df, df_team1, how ='left', on=['Tm'])
```

```python
#cast features to float type except the following three columns

dum = df.columns
dum = list(dum)
dum.remove('Player')
dum.remove('Pos')
dum.remove('Tm')

for header in dum:
    df[header] = df[header].astype(dtype='float')
```

```python
#drop players who were traded

df = df.set_index('Player')
x = df.groupby(['Player']).count() 
df = df[x.Tm == 1]
df = df.reset_index()
df = df.dropna()
df.Year = 2018
```

```python
#function to calculate the marginal PER for each player to the league's average that season.
#takes in 2 dataframe

def marginal_PER(player, season_data):
    avg_PER = np.mean(season_data.PER)
    marginal_PER = float(player.PER) - avg_PER
    return marginal_PER
```

```python
#looping through the dataframe to append their marginal PER as a feature
for idx, player in df.iterrows():
    df.loc[idx, 'M_PER'] = marginal_PER(player, df)
```

```python
#setting up dataset to be exported
df = df.set_index('Player')
df = df.drop('Tm', axis = 1)
df = pd.get_dummies(df)
df = df.drop('Pos_G', axis = 1)
df = df.rename(columns ={'PS/G':'PTS'})
```

```python
#exporting the dataframe as a csv file for testing

with open('NBA_2018_test.csv', 'w') as f:
    df.to_csv(f, header = True)
```
