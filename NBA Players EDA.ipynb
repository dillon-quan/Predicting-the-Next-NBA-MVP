```python
df_mvps = df[df.MVP == 1]
```

```python
df = df.drop('Year1', axis = 1)
```

```python
df_mvps.loc[:,['DBPM','OBPM']]
```

```python
from adjustText import adjust_text

fig, ax = plt.subplots()

ax.scatter(df_mvps['PER'], df_mvps['WS'], color= 'green')
plt.xlabel('PER')
plt.ylabel('WS')
plt.title('PER vs WS')

texts = []
for i, txt in enumerate(df_mvps.Year):
    texts.append(ax.text(df_mvps.PER[i], df_mvps.WS[i],txt))

adjust_text(texts)
plt.show()

#plotted the PER vs WS for each MVP. Each dot is labeled by the year. Labeling by the names gets too confusing to see.
```

```python
fig, ax = plt.subplots()

ax.scatter(df_mvps['DWS'], df_mvps['OWS'], color= 'red')
plt.xlabel('DWS')
plt.ylabel('OWS')
plt.title('DWS vs OWS')

texts = []
for i, txt in enumerate(df_mvps.Year):
    texts.append(ax.text(df_mvps.DWS[i], df_mvps.OWS[i],txt))

adjust_text(texts)
plt.show()
#Plotting the MVPs OBPM vs DBPM to show that great defense is not needed to win MVP
```

```python
plt.hist(df_mvps.Age, bins = 5, alpha = .5)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title("MVP Age Distribution")
plt.show()
#plotted a histogram to see what age range MVPs are won
```

```python
sns.swarmplot(x = 'Year', y = 'Pos', data = df_mvps)
plt.xlabel('Year')
plt.xticks(range(2000,2018, 2))
plt.ylabel('Position')
plt.title('MVP Position')
plt.show()
#a time series plot of the position that won the MVP
```

```python
player_career = pd.DataFrame()
mvp = []
#create a dataframe called player_career and a list called mvp

for i, player in df_mvps.iterrows():
    if player[1] in mvp:
        continue
    else:
        mvp.append(player[1])
        for j, names in df.iterrows():
            if player[1] == names[1]:
                player_career = pd.concat([player_career, df.loc[[j]]], axis = 0)
                
# a nested for loop to collect each MVP entire career stat lines. The if statement ensures that if a player wins multiple
#MVP, his career will not be plotted multiple times

        player_career = player_career.sort_values('Year', ascending = True)
        player_career.drop_duplicates(keep='first',inplace = True)
        
# This sorts drop any duplicates rows in the dataframe

        plt.figure(i)
        plt.plot(player_career.Year, player_career.PER, marker = 'o', color = 'green')
        plt.xlabel('Year')
        plt.ylabel('Player Efficiency Rating (PER)')
#Each player will have their own figure that plots the year vs PER

        plt.xticks(range(min(player_career.Year), max(player_career.Year), 2))
        #plt.yticks(range(0,#int(min(player_career.PER)), 
         #   int(max(player_career.PER)), 1)) 
        plt.title(player[1])
        
        player_career.reset_index()

        for k, series in player_career.iterrows():
            if series.MVP == 1:
                plt.annotate('MVP', 
                     xy = (player_career.loc[k,'Year'],player_career.loc[k,'PER']), 
                             xytext = (player_career.loc[k,'Year'] ,player_career.loc[k,'PER']+1),
                             arrowprops=dict(facecolor='black', shrink=.05, width = 1, headwidth = 1)
                             ,horizontalalignment='left', verticalalignment='top'
                      )
        #This for loop annotates the year each player won MVP
        plt.margins(.1)
        
    player_career = pd.DataFrame()
        #This clears the dataframe so each player can be plotted separately

plt.show()       
```
